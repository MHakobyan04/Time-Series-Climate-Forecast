# Time Series Climate Forecast
This notebook builds ARIMA and LSTM models to forecast temperature using the Daily Climate dataset. It demonstrates skills in time series analysis, Python, Pandas, PyTorch, and statistics for data-driven forecasting tasks.



```python
# Libraries imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
import torch
import torch.nn as nn
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Loading data
data_path  = 'DailyDelhiClimateTrain.csv'
try:
    data = pd.read_csv(data_path)
    print("Dataset loaded successfully.")
    print("First 5 rows:")
    print(data.head())
    print("\nDataset summary:")
    print(data.describe())
except Exception as e:
    print(f"Error loading dataset: {e}")
    print("Ensure DailyDelhiClimateTrain.csv is uploaded to Colab via Files > Upload.")
    raise

# Checking size
print(f"Total data points: {len(data)}")

# Temperature vizualization
try:
    plt.figure(figsize=(10, 5))
    plt.plot(data['date'], data['meantemp'], label='Mean Temperature')
    plt.title('Daily Mean Temperature in Delhi')
    plt.xlabel('Date')
    plt.ylabel('Temperature (°C)')
    plt.legend()
    plt.show()
except Exception as e:
    print(f"Error plotting data: {e}")
```

```python
# Transform data
try:
    data['date'] = pd.to_datetime(data['date'])
    data.set_index('date', inplace=True)
except Exception as e:
    print(f"Error converting date: {e}")
    raise

# Selecting the temperature
series = data['meantemp'].values
print(f"Series length: {len(series)}")

# Separating into train/test (80/20)
train_size = int(len(series) * 0.8)
test_size = len(series) - train_size
train, test = series[:train_size], series[train_size:]
print(f"Train size: {train_size}, Test size: {test_size}")

# Checking
if len(train) + len(test) != len(series):
    raise ValueError("Train/test split error")
```

```python
# ARIMA model
try:
    arima_model = ARIMA(train, order=(5, 1, 0))
    arima_fit = arima_model.fit()
    arima_pred = arima_fit.forecast(steps=len(test))
except Exception as e:
    print(f"ARIMA error: {e}")
    raise

# Metrics of ARIMA
arima_mae = mean_absolute_error(test, arima_pred)
arima_rmse = np.sqrt(mean_squared_error(test, arima_pred))
print(f'ARIMA MAE: {arima_mae:.4f}')
print(f'ARIMA RMSE: {arima_rmse:.4f}')

# ARIMA vizualization
plt.figure(figsize=(10, 5))
plt.plot(test, label='Actual')
plt.plot(arima_pred, label='ARIMA Prediction')
plt.title('ARIMA Forecast')
plt.legend()
plt.show()
```

```python
# Preparing data for LSTM
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train.reshape(-1, 1))
test_scaled = scaler.transform(test.reshape(-1, 1))

# Creating a sequences (size = 10)
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

seq_length = 10
X_train, y_train = create_sequences(train_scaled, seq_length)
X_test, y_test = create_sequences(test_scaled, seq_length)
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
# Transform into tensors
X_train = torch.FloatTensor(X_train)
y_train = torch.FloatTensor(y_train)
X_test = torch.FloatTensor(X_test)
y_test = torch.FloatTensor(y_test)

# Definition of LSTM model
class LSTM(nn.Module):
    def __init__(self, input_size=1, hidden_size=50, num_layers=1):
        super(LSTM, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        out, _ = self.lstm(x)
        out = self.fc(out[:, -1, :])
        return out

# LSTM initialization
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = LSTM().to(device)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
print(f"LSTM will run on: {device}")

# Fitting LSTM
epochs = 20
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(X_train.to(device))
    loss = criterion(outputs, y_train.to(device))
    loss.backward()
    optimizer.step()
    print(f'Epoch {epoch+1}, Loss: {loss.item():.4f}')

# LSTM prediction
model.eval()
with torch.no_grad():
    lstm_pred = model(X_test.to(device))
lstm_pred = scaler.inverse_transform(lstm_pred.cpu().numpy())
y_test_inv = scaler.inverse_transform(y_test.cpu().numpy())

# Checking lenghts
print(f"Length of y_test_inv: {len(y_test_inv)}")
print(f"Length of lstm_pred: {len(lstm_pred)}")
if len(y_test_inv) != len(lstm_pred):
    raise ValueError("Inconsistent lengths in LSTM predictions")

# LSTM's metrics
lstm_mae = mean_absolute_error(y_test_inv, lstm_pred)
lstm_rmse = np.sqrt(mean_squared_error(y_test_inv, lstm_pred))
print(f'LSTM MAE: {lstm_mae:.4f}')
print(f'LSTM RMSE: {lstm_rmse:.4f}')

# LSTM vizualization
plt.figure(figsize=(10, 5))
plt.plot(y_test_inv, label='Actual')
plt.plot(lstm_pred, label='LSTM Prediction')
plt.title('LSTM Forecast')
plt.legend()
plt.show()
```
